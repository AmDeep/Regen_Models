```python
file_reg.loc[file_reg['sex'] == 0, 'sex'] = 'Female'
```

```python
file_reg.loc[file_reg['sex'] == 1, 'sex'] = 'Male'
```

```python
file_reg.loc[file_reg['fbs'] == 0, 'fbs'] = 'Normal fbs'
```

```python
file_reg.loc[file_reg['fbs'] == 1, 'fbs'] = 'High fbs'
```

```python
file_reg.loc[file_reg['restecg'] == 0, 'restecg'] = 'Probable'
```

```python
file_reg.loc[file_reg['restecg'] == 1, 'restecg'] = 'Normal'
```

```python
file_reg.loc[file_reg['restecg'] == 2, 'restecg'] = 'Abnormal'
```

```python
file_reg.loc[file_reg['exang'] == 0, 'exang'] = 'Not EIA'
file_reg.loc[file_reg['exang'] == 1, 'exang'] = 'EIA'
```

```python
file_reg.loc[file_reg['thal'] == 1, 'thal'] = 'Fixed Defect'
file_reg.loc[file_reg['thal'] == 2, 'thal'] = 'Normal thal'
file_reg.loc[file_reg['thal'] == 3, 'thal'] = 'Reversible Defect'
```

```python
file_reg[file_reg['cp'] = 0, 'Asymptomatic'] = 'True' 
```

```python
one_hot=pd.get_dummies(file_reg['cp'])
# Drop column as it is now encoded
file_reg = file_reg.drop('cp',axis = 1)
# Join the encoded df
file_reg = file_reg.join(one_hot)
```

```python
two_hot=pd.get_dummies(file_reg['fbs'])
# Drop column as it is now encoded
file_reg = file_reg.drop('fbs',axis = 1)
# Join the encoded df
file_reg = file_reg.join(two_hot)
```

```python
three_hot=pd.get_dummies(file_reg['sex'])
# Drop column as it is now encoded
file_reg = file_reg.drop('sex',axis = 1)
# Join the encoded df
file_reg = file_reg.join(three_hot)
```

```python
four_hot=pd.get_dummies(file_reg['restecg'])
# Drop column as it is now encoded
file_reg = file_reg.drop('restecg',axis = 1)
# Join the encoded df
file_reg = file_reg.join(four_hot)
```

```python
five_hot=pd.get_dummies(file_reg['exang'])
# Drop column as it is now encoded
file_reg = file_reg.drop('exang',axis = 1)
# Join the encoded df
file_reg = file_reg.join(five_hot)
```

```python
six_hot=pd.get_dummies(file_reg['thal'])
# Drop column as it is now encoded
file_reg = file_reg.drop('thal',axis = 1)
# Join the encoded df
file_reg = file_reg.join(six_hot)
```
